```python
from google.colab.patches import cv2_imshow
import cv2

# Reading the video
vid = cv2.VideoCapture("sunset.mp4")

while True:
    isTrue, frame = vid.read()
    if not isTrue:
        break

    # Displaying the frame
    cv2_imshow(frame)

    # Wait for a short period to simulate video playback
    cv2.waitKey(20)

# Release video capture and close windows
vid.release()
cv2.destroyAllWindows()
```

```python
from google.colab.patches import cv2_imshow
import cv2
import time

# Reading the video
vid = cv2.VideoCapture("sunset.mp4")

# Desired width and height for resizing
width, height = 680, 440

while True:
    isTrue, frame = vid.read()
    if not isTrue:
        break

    # Resize the frame
    frame_resized = cv2.resize(frame, (width, height))

    # Display the resized frame
    cv2_imshow(frame_resized)

    # Simulate video playback speed
    time.sleep(0.03)  # Adjust the delay for playback speed

# Release video capture
vid.release()
```

```python
blank=np.zeros((500,500,3), dtype=np.uint8)
cv2_imshow(blank)
cv2.waitKey(60)
```

```python
## box in blank
blank[200:300, 300:400]=0,0,255
cv2_imshow(blank)
cv2.waitKey(60)
```

```python
## draw rectangle
cv2.rectangle(blank, (0,0), (250, 250), (0,255,0) , thickness=2)
cv2_imshow(blank)
cv2.waitKey(60)
```
